```python
 #Step 1: Import Libraries
 # Data manipulation
import pandas as pd
import numpy as np

# Text processing
from textblob import TextBlob
import re
import string

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Settings
sns.set(style="whitegrid")
```

```python
#Step 2: Load and Prepare Data
# Load the dataset
df = pd.read_csv("chatgpt_reviews.csv")

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Display first few rows
df.head()
```

```python
#Step 3: Data Cleaning
# Check for missing values
df.isnull().sum()

# Fill missing reviews with empty string
df['review'] = df['review'].fillna("")

# Convert review_date to datetime
df['review_date'] = pd.to_datetime(df['review_date'], errors='coerce')

# Ensure ratings are numeric
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Drop rows with missing dates or ratings if necessary
df.dropna(subset=['review_date', 'ratings'], inplace=True)

# Check data types
df.dtypes
```

```python
### 📊 Step 4: Exploratory Data Analysis (EDA)
plt.figure(figsize=(8, 5))
sns.countplot(x='ratings', data=df, palette='viridis')
plt.title("Distribution of Ratings")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```

```python
#Review Length Distribution
df['review_length'] = df['review'].apply(lambda x: len(x.split()))

plt.figure(figsize=(10, 5))
sns.histplot(df['review_length'], bins=50, kde=True)
plt.title("Review Length Distribution")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

```python
#Reviews Over Time
df['date'] = df['review_date'].dt.date
daily_reviews = df.groupby('date').size()

plt.figure(figsize=(12, 6))
daily_reviews.plot()
plt.title("Number of Reviews Over Time")
plt.xlabel("Date")
plt.ylabel("Review Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
###  Step 5: Sentiment Analysis
# Apply TextBlob for sentiment polarity and subjectivity
df['polarity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Categorize sentiment
def get_sentiment(p):
    if p > 0.1:
        return 'Positive'
    elif p < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['polarity'].apply(get_sentiment)

# Sentiment distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='sentiment', data=df, palette='Set2')
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
# Clean Text
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\\S+|www\\S+|https\\S+", '', text)
    text = re.sub(r'[^A-Za-z0-9 ]+', '', text)
    text = re.sub(r'\\s+', ' ', text).strip()
    return text

df['cleaned_review'] = df['review'].apply(clean_text)
```

```python
#Word Cloud for Positive Reviews
positive_reviews = df[df['sentiment'] == 'Positive']['cleaned_review']
text = " ".join(positive_reviews)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Positive Reviews")
plt.show()
```

```python
### Step 7: Sentiment vs Ratings
plt.figure(figsize=(10, 6))
sns.boxplot(x='ratings', y='polarity', data=df)
plt.title("Sentiment Polarity vs Ratings")
plt.xlabel("Ratings")
plt.ylabel("Polarity")
plt.show()
```
